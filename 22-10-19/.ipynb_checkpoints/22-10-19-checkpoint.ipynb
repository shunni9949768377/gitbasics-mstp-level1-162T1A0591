```python
import matplotlib.pyplot as a
x=[2,4]
y=[2,0]
a.plot(x,y)
a.xlabel('x-Data')
a.ylabel('y-axis')
a.title('line plot')
a.grid()
a.show()
```

# bargraph

```python
import matplotlib.pyplot as a
x=[2,4]
y=[2,0]
a.bar(x,y)
a.xlabel('x-Data')
a.ylabel('y-axis')
a.title('bar plot')
a.grid()
a.show()
```

# histogram

```python
import matplotlib.pyplot as a
x=list(range(1,10,3))
y=[3,5,7,9]
a.hist(x,y)
a.xlabel('x-Data')
a.ylabel('y-axis')
a.title('hist plot')
a.grid()
a.show()
```

```python
import matplotlib.pyplot as a
distance=list(range(1,10,3))
time=[3,5,7,9]
distance=2*time
a.hist(distance,time)
a.xlabel('x-Data')
a.ylabel('y-axis')
a.title('hist plot')
a.grid()
a.show()
```

# Ticks

```python
import matplotlib.pyplot as a
x=list(range(5,51,5))
y=list(range(1,11))
a.plot(x,y)
a.title("a graph with example of ticks")
a.xlabel("natural numbers")
a.ylabel("values with ++5 upto 50")
a.show
```

```python
import matplotlib.pyplot as a
y=[5,10,15,20,25,30,35,40,45,50]
print(x)
x=[9,8,5,6,3,2,1,4,7,10]
print(y)
a.plot(x,y)
a.title("a graph with example of ticks")
a.xlabel("natural numbers")
a.ylabel("values with ++5 upto 50")
a.show
```

```python
import matplotlib.pyplot as a
import numpy as np
a.xticks(np.arange(1,11))
a.yticks(np.arange(5,51,5))
a.plot(x,y)
a.title("matplot-Ticks ")
a.xlabel("10 natural numbers")
a.ylabel("1-10 values with 5+")
a.show()
```

```python
import numpy as np
import matplotlib.pyplot as pl
T=np.arange(1,10,0.1)
Amplitude=np.sin(T)
pl.plot(T,Amplitude)
pl.title("sine wave")
pl.xlabel("Time")
pl.ylabel("Amplitude")
pl.show()
```

```python
import pandas as pd
df=pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as pl
years=df.columns[2::]
income=df.iloc[1,2:]
pl.plot(years,income)
pl.title("income of alaska from 2005-2013")
pl.xlabel("years")
pl.ylabel("income of alaska")
pl.savefig("income of alaska.jpg")
pl.show()
```

```python
import matplotlib.pyplot as pl
years=df.columns[-5::]
income=df.iloc[0:,2:]
pl.plot(years,income)
pl.title("income of states from 2005-2013")
pl.xlabel("years")
pl.ylabel("income of states")
pl.savefig("income of states.jpg")
pl.show()
```
